```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file = 'MLS_December_2018__Content_Posts_fixed.xls'

xl = pd.ExcelFile(file)

df = xl.parse('Posts Report')
```

```python
df.groupby('Page')['Fans'].mean().sort_values(ascending=False).plot.bar()
```

```python
df.groupby('Page')['Total Actions % Fans'].mean().sort_values(ascending=False).plot.bar()
```

```python
df.groupby('Post Type')['Total Actions % Fans'].mean().sort_values(ascending=False)
```

```python
df.groupby('Post Type')['Total Actions'].mean().sort_values(ascending=False)
```

```python
var = 'Page'
data = pd.concat([df['Total Actions % Fans'], df[var]], axis=1)
f, ax = plt.subplots(figsize=(12,8))
fig = sns.barplot(x=var, y='Total Actions % Fans', data=data)
fig.axis(ymin=0, ymax=.01);
xt = plt.xticks(rotation=90)
```

```python
df['Created Time'] = pd.to_datetime(df['Created Time'])
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(20, 8)
fig.autofmt_xdate()
x = df.groupby('Created Time')['Total Actions % Fans'].mean()
ax.plot(x)
ax.set_ylabel('Average Total Actions % Fans')
plt.show()
```

```python
var = 'Page'
data = pd.concat([df['Shares Count'], df[var]], axis=1)
f, ax = plt.subplots(figsize=(12,8))
fig = sns.boxplot(x=var, y='Shares Count', data=data)
fig.axis(ymin=0, ymax=400);
xt = plt.xticks(rotation=90)
```

```python
var = 'Word Count'
data = pd.concat([df['Total Actions % Fans'], df[var]], axis=1)
ax = sns.barplot(x=var, y='Total Actions % Fans', data=data)
xt = plt.xticks(rotation=45)
```

```python
var = 'Post Type'
data = pd.concat([df['Comments Count'], df[var]], axis=1)
ax = sns.barplot(x=var, y='Comments Count', data=data, estimator=np.mean)
xt = plt.xticks(rotation=45)
```
